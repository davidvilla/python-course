# Python

Python es un lenguaje interactivo, es decir, además ejecutar programas contenidos en un fichero de código fuente, permite al programador introducir nuevas sentencias una a una, como en una shell. Esto es de gran utilidad en muchas situaciones, pero especialmente en el aprendizaje del lenguaje. Para utilizar el intérprete de forma interactiva, simplemente ejecuta:

A partir de ese momento, el programador puede introducir sentencias Python, que serán ejecutadas inmediatamente (como las que aparecen en este documento).

Python es un lenguaje dinámico, interpretado, orientado a objetos y extraordinariamente versatil. Empecemos, como siempre, con el típico «hola mundo».

```python
print("Hello world")
```

En la sentencia anterior, **print** es una función y el texto entre comillas es una cadena de texto (del tipo **str**) para ser más exactos. Cuando se utiliza un valor directamente como en este caso, se dice que se trata de un «literal». La alternativa es utilizar una variable para almacenarlo.

# Variables

```python
text = "Hello World"
print(text)
```

La variable **text** es de tipo cadena. Todos los tipos en Python son clases porque todo es orientado a objetos. Se puede obtener el tipo de cualquier variable con la función **type()**

```python
type(text)
```

Y se puede averiguar los métodos de un objeto con la función **dir()**

```python
print(dir(text))
```

También dispone de un sistema de ayuda en línea:

```python
help(str)
```

Por ejemplo, el método **lower()** devuelve una *nueva* cadena con el mismo texto que el objeto original, pero todo en minúsculas. Es importante señalar que la cadena original permanece inalterada.

```python
print(text.lower())
print(text)
```
